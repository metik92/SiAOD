```python
def incidence_to_adjacency(mtrx):
    v = len(mtrx)
    e = len(mtrx[0]) 
    

    adj_matrix = [[0 for _ in range(v)] for _ in range(v)]
    
    for edge_idx in range(e):
        start = end = None
        weight = 0
        
        for vertex_idx in range(v):
            val = mtrx[vertex_idx][edge_idx]
            
            if val > 0:
                start = vertex_idx
                weight = val
            elif val < 0:
                end = vertex_idx
                
        if start is not None and end is not None:
            adj_matrix[start][end] = weight
            
    return adj_matrix
```

```python
def floyd_warshall(graph,start,end):
    n = len(graph)
    INF = float('inf')
    
    # Подготовка матрицы
    dist = [
        [INF if i != j and graph[i][j] == 0 else graph[i][j] for j in range(n)]
        for i in range(n)
    ]
    
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    
    return dist[start][end] if dist[start][end] != INF else None

```

```python
def dijkstra(graph, start, end):
    n = len(graph)
    INF = float('inf')
    dist = [INF] * n
    visited = [False] * n
    dist[start] = 0

    for _ in range(n):
        u = -1
        min_dist = INF
        for i in range(n):
            if not visited[i] and dist[i] < min_dist:
                min_dist = dist[i]
                u = i

        if u == -1:  
            break

        visited[u] = True

        for v in range(n):
            weight = graph[u][v]
            if weight > 0 and not visited[v]:
                if dist[u] + weight < dist[v]:
                    dist[v] = dist[u] + weight

    return dist[end] if dist[end] != INF else None
```

```python
def bellman_ford(graph, start, end):
    V = len(graph)
    INF = float('inf')
    dist = [INF] * V  
    dist[start] = 0  

    for _ in range(V - 1): 
        for i in range(V):
            for j in range(V):
                if graph[i][j] != 0 and dist[i] + graph[i][j] < dist[j]:
                    dist[j] = dist[i] + graph[i][j]
 
  
    for i in range(V):
        for j in range(V):
            if graph[i][j] != 0 and dist[i] + graph[i][j] < dist[j]:
                return None  

    return dist[end] if dist[end] != INF else None
graph = [
    [0, 3, 0, 5],
    [2, 0, 0, 4],
    [0, 1, 0, 0],
    [0, 0, 2, 0]]

print(bellman_ford(graph,0,2))
```

```python
import heapq
INF = float('inf')
def bellman_ford(graph, src):
    V = len(graph)
    dist = [INF] * V
    dist[src] = 0

    for _ in range(V):
        for u in range(V):
            for v, weight in enumerate(graph[u]):
                if weight != INF and dist[u] + weight < dist[v]:
                    dist[v] = dist[u] + weight

    return dist


def dijkstra(graph, start, adj):
    V = len(graph)
    dist = [INF] * V
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, weight in adj[u]:
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                heapq.heappush(pq, (dist[v], v))

    return dist

def johnson(graph):
    
    V = len(graph)
    for i in range(V):
        graph.append([0] * V)  

    
    h = bellman_ford(graph, V)

    new_adj = [[] for _ in range(V)]
    for u in range(V):
        for v in range(V):
            if graph[u][v] != INF:
                new_w = graph[u][v] + h[u] - h[v]
                new_adj[u].append((v, new_w))

    dist_matrix = []
    for u in range(V):
        dist_u = dijkstra(graph, u, new_adj)
        dist_u = [d - h[u] + h[v] if d != INF else INF for v, d in enumerate(dist_u)]
        dist_matrix.append(dist_u)

    return dist_matrix
```

```python
from collections import deque
def levit(graph, start):
    V = len(graph)
    INF = float('inf')
    dist = [INF] * V
    prev = [None] * V
    m0, m1, m2 = set(), deque(), set()

    dist[start] = 0
    m1.append(start)

    while m1:
        u = m1.popleft()
        m0.discard(u)

        for v in range(V):
            w = graph[u][v]
            if w == INF or u == v:
                continue

            if v not in m0 and v not in m1 and v not in m2:
                dist[v] = dist[u] + w
                prev[v] = u
                m1.append(v)

            elif v in m1 and dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                prev[v] = u

            elif v in m2 and dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                prev[v] = u
                m1.appendleft(v)
                m2.remove(v)

        m2.add(u)

    return dist, prev
```

```python
from copy import *
def dijkstra_path(graph, start, end):
    V = len(graph)
    dist = [INF] * V
    prev = [None] * V
    dist[start] = 0
    queue = [(0, start)]

    while queue:
        d, u = heapq.heappop(queue)
        if u == end:
            break
        if d > dist[u]:
            continue
        for v, weight in enumerate(graph[u]):
            if weight != INF and dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                prev[v] = u
                heapq.heappush(queue, (dist[v], v))

    path = []
    node = end
    while node is not None:
        path.append(node)
        node = prev[node]
    return path[::-1], dist[end]

# --- Алгоритм Йена ---
def yen_k_shortest_paths(graph, start, end, k):
    A = []
    B = []

    path, cost = dijkstra_path(graph, start, end)
    if not path or cost == INF:
        return []

    A.append((cost, path))

    for i in range(1, k):
        for j in range(len(A[i-1][1]) - 1):
            spur_node = A[i-1][1][j]
            root_path = A[i-1][1][:j+1]

            g_copy = deepcopy(graph)

            for cost_path, p in A:
                if p[:j+1] == root_path and len(p) > j+1:
                    u, v = p[j], p[j+1]
                    g_copy[u][v] = INF

            for node in root_path[:-1]:
                for v in range(len(g_copy)):
                    g_copy[node][v] = INF

            spur_path, spur_cost = dijkstra_path(g_copy, spur_node, end)
            if spur_path and spur_cost != INF:
                total_path = root_path[:-1] + spur_path
                total_cost = sum(graph[total_path[i]][total_path[i+1]] for i in range(len(total_path) - 1))
                if (total_cost, total_path) not in B:
                    B.append((total_cost, total_path))

        if not B:
            break

        B.sort()
        A.append(B.pop(0))

    return A
```

```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import ast
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class GraphApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Анализатор графов")
        self.root.geometry("900x700")
        
        # Инициализация графа
        self.G = nx.DiGraph()
        self.graph_data = None
        self.figure = plt.figure(figsize=(8, 6))
        
        # Создание интерфейса
        self.create_widgets()
    
    def create_widgets(self):
        # Фрейм для ввода данных
        input_frame = ttk.LabelFrame(self.root, text="Ввод графа", padding=10)
        input_frame.pack(fill=tk.BOTH, expand=True, padx=10, pady=5)
        
        # Выбор типа матрицы
        ttk.Label(input_frame, text="Тип матрицы:").grid(row=0, column=0, sticky=tk.W)
        self.matrix_type = tk.StringVar(value="adjacency")
        ttk.Radiobutton(input_frame, text="Матрица смежности", 
                       variable=self.matrix_type, value="adjacency").grid(row=0, column=1, sticky=tk.W)
        ttk.Radiobutton(input_frame, text="Матрица инцидентности", 
                       variable=self.matrix_type, value="incidence").grid(row=0, column=2, sticky=tk.W)
        
        # Поле для ввода матрицы
        self.text_input = tk.Text(input_frame, height=10, width=50, font=('Courier New', 10))
        self.text_input.grid(row=1, column=0, columnspan=3, pady=5)
        
        # Кнопки управления
        ttk.Button(input_frame, text="Пример матрицы", command=self.insert_example).grid(row=2, column=0, pady=5)
        ttk.Button(input_frame, text="Загрузить из файла", command=self.load_from_file).grid(row=2, column=1, pady=5)
        ttk.Button(input_frame, text="Принять граф", command=self.accept_graph).grid(row=2, column=2, pady=5)
        
        # Фрейм для анализа
        analysis_frame = ttk.LabelFrame(self.root, text="Анализ графа", padding=10)
        analysis_frame.pack(fill=tk.BOTH, expand=True, padx=10, pady=5)
        
        # Выбор алгоритма
        ttk.Label(analysis_frame, text="Алгоритм:").grid(row=0, column=0, sticky=tk.W)
        self.algorithm = tk.StringVar(value="dijkstra")
        algorithms = [("Дейкстры", "dijkstra"), ("Флойда-Уоршелла", "floyd"),
                     ("Беллмана-Форда", "bellman"), ("Джонсона", "johnson")]
        
        for i, (text, val) in enumerate(algorithms):
            ttk.Radiobutton(analysis_frame, text=text, variable=self.algorithm, 
                           value=val).grid(row=0, column=i+1, sticky=tk.W)
        
        # Ввод вершин
        ttk.Label(analysis_frame, text="Стартовая вершина:").grid(row=1, column=0, sticky=tk.W)
        self.start_node = ttk.Entry(analysis_frame, width=5)
        self.start_node.grid(row=1, column=1, sticky=tk.W)
        
        ttk.Label(analysis_frame, text="Конечная вершина:").grid(row=1, column=2, sticky=tk.W)
        self.end_node = ttk.Entry(analysis_frame, width=5)
        self.end_node.grid(row=1, column=3, sticky=tk.W)
        
        ttk.Button(analysis_frame, text="Найти путь", command=self.find_path).grid(row=1, column=4, padx=5)
        ttk.Button(analysis_frame, text="Визуализировать", command=self.visualize_graph).grid(row=1, column=5, padx=5)
        
        # Фрейм для результатов
        result_frame = ttk.LabelFrame(self.root, text="Результаты", padding=10)
        result_frame.pack(fill=tk.BOTH, expand=True, padx=10, pady=5)
        
        # Текстовое поле для результатов
        self.result_text = tk.Text(result_frame, height=5, width=80)
        self.result_text.pack(fill=tk.BOTH, expand=True)
        
        # Холст для визуализации графа
        self.canvas = FigureCanvasTkAgg(self.figure, master=result_frame)
        self.canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
    
    def insert_example(self):
        """Вставляет пример матрицы в поле ввода"""
        example = """# Пример матрицы смежности (3 вершины)
[[0, 2, 0],
 [0, 0, 3],
 [0, 0, 0]]

# Пример матрицы инцидентности (3 вершины, 2 ребра)
[[1, 0],
 [1, 1],
 [0, 1]]"""
        self.text_input.delete(1.0, tk.END)
        self.text_input.insert(tk.END, example)
    
    def load_from_file(self):
        """Загружает матрицу из файла"""
        filepath = filedialog.askopenfilename(filetypes=[("Text files", "*.txt"), ("All files", "*.*")])
        if filepath:
            try:
                with open(filepath, 'r') as file:
                    content = file.read()
                    self.text_input.delete(1.0, tk.END)
                    self.text_input.insert(tk.END, content)
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить файл: {e}")
    
    def accept_graph(self):
        """Принимает и проверяет введённые данные графа"""
        try:
            text = self.text_input.get(1.0, tk.END).strip()
            
            # Удаляем комментарии (строки, начинающиеся с #)
            cleaned_lines = []
            for line in text.split('\n'):
                if not line.strip().startswith('#'):
                    cleaned_lines.append(line)
            cleaned_text = '\n'.join(cleaned_lines)
            
            self.graph_data = ast.literal_eval(cleaned_text)
            
            if not isinstance(self.graph_data, list) or not all(isinstance(row, list) for row in self.graph_data):
                raise ValueError("Ввод должен быть матрицей (списком списков)")
            
            if self.matrix_type.get() == "adjacency":
                if len(self.graph_data) != len(self.graph_data[0]):
                    raise ValueError("Матрица смежности должна быть квадратной")
                self.G = nx.DiGraph()
                for i in range(len(self.graph_data)):
                    for j in range(len(self.graph_data[i])):
                        if self.graph_data[i][j] != 0:
                            self.G.add_edge(i, j, weight=self.graph_data[i][j])
            else:
                # Преобразование матрицы инцидентности в граф
                num_nodes = len(self.graph_data)
                num_edges = len(self.graph_data[0]) if num_nodes > 0 else 0
                self.G = nx.DiGraph()
                
                for j in range(num_edges):
                    edge = []
                    for i in range(num_nodes):
                        if self.graph_data[i][j] != 0:
                            edge.append((i, self.graph_data[i][j]))
                    
                    if len(edge) == 2:
                        self.G.add_edge(edge[0][0], edge[1][0], weight=edge[1][1])
            
            messagebox.showinfo("Успех", f"Граф успешно загружен!\nВершин: {len(self.G.nodes())}\nРёбер: {len(self.G.edges())}")
        except Exception as e:
            messagebox.showerror("Ошибка", f"Некорректные данные графа: {e}")
    
    def visualize_graph(self):
        """Визуализирует граф"""
        if len(self.G.nodes()) == 0:
            messagebox.showerror("Ошибка", "Граф не загружен или пуст!")
            return
        
        self.figure.clear()
        pos = nx.spring_layout(self.G)
        nx.draw(self.G, pos, with_labels=True, node_color='skyblue', 
                node_size=500, arrows=True, ax=self.figure.add_subplot(111))
        labels = nx.get_edge_attributes(self.G, 'weight')
        nx.draw_networkx_edge_labels(self.G, pos, edge_labels=labels)
        self.canvas.draw()
    
    def find_path(self):
        """Находит кратчайший путь между вершинами"""
        if len(self.G.nodes()) == 0:
            messagebox.showerror("Ошибка", "Граф не загружен!")
            return
        
        try:
            start = int(self.start_node.get())
            end = int(self.end_node.get())
            
            if start not in self.G.nodes() or end not in self.G.nodes():
                raise ValueError("Номера вершин должны быть в диапазоне 0..{}".format(len(self.G.nodes())-1))
            
            algorithm = self.algorithm.get()
            result = ""
            
            if algorithm == "dijkstra":
                try:
                    path = nx.dijkstra_path(self.G, start, end)
                    length = nx.dijkstra_path_length(self.G, start, end)
                    result = f"Алгоритм Дейкстры\nПуть: {path}\nДлина: {length}"
                    self.highlight_path(path)
                except nx.NetworkXNoPath:
                    result = "Путь между вершинами не существует"
            
            elif algorithm == "floyd":
                try:
                    path = nx.floyd_warshall_predecessor_and_distance(self.G)
                    predecessors = path[0]
                    current = end
                    path_nodes = [current]
                    while current != start:
                        current = predecessors[current][start]
                        path_nodes.append(current)
                    path_nodes.reverse()
                    distance = nx.floyd_warshall_numpy(self.G)[start][end]
                    result = f"Алгоритм Флойда-Уоршелла\nПуть: {path_nodes}\nДлина: {distance}"
                    self.highlight_path(path_nodes)
                except:
                    result = "Ошибка при выполнении алгоритма Флойда-Уоршелла"
            
            self.result_text.delete(1.0, tk.END)
            self.result_text.insert(tk.END, result)
            
        except ValueError as e:
            messagebox.showerror("Ошибка", str(e))
    
    def highlight_path(self, path):
        """Подсвечивает найденный путь на графе"""
        if not path:
            return
        
        self.figure.clear()
        pos = nx.spring_layout(self.G)
        nx.draw(self.G, pos, with_labels=True, node_color='skyblue', 
                node_size=500, arrows=True, ax=self.figure.add_subplot(111))
        
        # Рисуем рёбра пути красным цветом
        path_edges = list(zip(path[:-1], path[1:]))
        nx.draw_networkx_edges(self.G, pos, edgelist=path_edges, 
                              edge_color='r', width=2, ax=self.figure.gca())
        
        labels = nx.get_edge_attributes(self.G, 'weight')
        nx.draw_networkx_edge_labels(self.G, pos, edge_labels=labels)
        self.canvas.draw()

if __name__ == "__main__":
    root = tk.Tk()
    app = GraphApp(root)
    root.mainloop()
```
