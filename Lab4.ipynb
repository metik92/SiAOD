```python
#бинарный поиск 

def binary_search(arr, num):
    low = 0 
    high = len(arr) - 1 
    while low <= high: 
        mid = (low + high) // 2
        if arr[mid] == num: 
            return mid
        elif arr[mid] > num:
            high = mid - 1
        else:
            low = mid + 1
    return None 
```

```python
def fibonacci_search(arr, item):
    # Определяем числа Фибоначчи
    fib1, fib2 = 0, 1
    fib_sum = fib1 + fib2
    while fib_sum < len(arr):
        fib1 = fib2
        fib2 = fib_sum
        fib_sum = fib1 + fib2
    offset = -1
    while fib_sum > 1:
        i = min(offset + fib1, len(arr) - 1)
        if arr[i] < item:
            fib_sum = fib2
            fib2 = fib1
            fib1 = fib_sum - fib2
            offset = i
        elif arr[i] > item:
            fib_sum = fib1
            fib2 = fib2 - fib1
            fib1 = fib_sum - fib2
        else:
            return i
    if fib2 == 1 and arr[offset+1] == item:
        return offset + 1
    return None
```

```python
def interpolation_search(arr, item):
    low = 0
    high = len(arr) - 1
    while low <= high and item >= arr[low] and item <= arr[high]:
        pos = low + ((item - arr[low]) * (high - low) // (arr[high] - arr[low])) 
        if arr[pos] == item: 
            return pos
        elif arr[pos] < item:
            low = pos + 1
        else: 
            high = pos - 1
    return None 
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key      # Ключ элемента
        self.value = value  # Значение элемента
        self.is_deleted = False  # Флаг удаления (для линейного пробирования)

class HashTable_linear:
    def __init__(self):
        self.capacity = 10  # Начальный размер таблицы
        self.size = 0       # Количество элементов
        self.table = [None] * self.capacity  # Массив Node объектов

    def _hash(self, key):
     
        return hash(key) % self.capacity

    def _rehash(self, old_hash):
      
        return (old_hash + 1) % self.capacity

    def add(self, key, value):
        
        if self.size >= 0.7 * self.capacity:
            self._resize()

        hash_idx = self._hash(key)
        start_idx = hash_idx

        while True:
            # Если ячейка пуста или удалена
            if self.table[hash_idx] is None or self.table[hash_idx].is_deleted:
                self.table[hash_idx] = Node(key, value)
                self.size += 1
                return
            
            # Если ключ уже существует
            elif self.table[hash_idx].key == key:
                self.table[hash_idx].value = value
                return
            
            # Линейное пробирование
            hash_idx = self._rehash(hash_idx)
            
            # Если прошли всю таблицу
            if hash_idx == start_idx:
                self._resize()
                self.add(key, value)
                return

    def get(self, key):
    
        hash_idx = self._hash(key)
        start_idx = hash_idx

        while True:
            node = self.table[hash_idx]
            
            # Если нашли существующий ключ
            if node is not None and not node.is_deleted and node.key == key:
                return node.value
            
            # Если пустая ячейка (значит ключа нет)
            elif node is None:
                return None
            
            # Продолжаем поиск
            hash_idx = self._rehash(hash_idx)
            
            # Если прошли всю таблицу
            if hash_idx == start_idx:
                return None

    def remove(self, key):
        
        hash_idx = self._hash(key)
        start_idx = hash_idx

        while True:
            node = self.table[hash_idx]
            
            # Если нашли существующий ключ
            if node is not None and not node.is_deleted and node.key == key:
                node.is_deleted = True
                self.size -= 1
                return
            
            # Если пустая ячейка (значит ключа нет)
            elif node is None:
                return
            
            # Продолжаем поиск
            hash_idx = self._rehash(hash_idx)
            
            # Если прошли всю таблицу
            if hash_idx == start_idx:
                return

    def _resize(self):
       
        old_table = self.table
        old_capacity = self.capacity

        self.capacity *= 2
        self.table = [None] * self.capacity
        self.size = 0

        # Перехешируем все существующие элементы
        for node in old_table:
            if node is not None and not node.is_deleted:
                self.add(node.key, node.value)

ht = HashTable_linear()
ht.add("apple", 5)
ht.add("banana", 10)
ht.add("orange", 7)
ht.add("apple", 6)  

start_time = time.time()
ht.get("apple")
print("линейное рехэширование:--- {0}  ---".format((time.time() - start_time)))
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class HashTable_ceoi4ka:
    def __init__(self):
        self.capacity = 10 # начальная емкость таблицы
        self.size = 0 # количество элементов в таблице
        self.buckets = [None] * self.capacity # создаем пустые списки для каждой ячейки таблицы
  
    # Метод добавления элемента в таблицу
    def add(self, key, value):
        index = hash(key) % self.capacity # вычисляем хэш-код ключа и индекс ячейки таблицы
        node = self.buckets[index] # получаем список (цепочку) элементов по индексу
        while node: # проходим по всем элементам цепочки
            if node.key == key: # если ключ уже есть в таблице, обновляем значение
                node.value = value
                return
            node = node.next
        # если ключа нет в таблице, добавляем его в начало цепочки
        new_node = Node(key, value)
        new_node.next = self.buckets[index]
        self.buckets[index] = new_node
        self.size += 1
        # если количество элементов превысило 70% от емкости таблицы, увеличиваем емкость вдвое
        if self.size > 0.7 * self.capacity:
            self.resize()
  
    # Метод получения значения по ключу
    def get(self, key):
        index = hash(key) % self.capacity # вычисляем хэш-код ключа и индекс ячейки таблицы
        node = self.buckets[index] # получаем список (цепочку) элементов по индексу
        while node: # проходим по всем элементам цепочки
            if node.key == key: # если ключ найден, возвращаем его значение
                return node.value
            node = node.next
        # если ключ не найден, возвращаем None
        return None
  
    # Метод удаления элемента из таблицы
    def remove(self, key):
        index = hash(key) % self.capacity # вычисляем хэш-код ключа и индекс ячейки таблицы
        node = self.buckets[index] # получаем список (цепочку) элементов по индексу
        prev = None
        while node: # проходим по всем элементам цепочки
            if node.key == key: # если ключ найден, удаляем элемент
                if prev:
                    prev.next = node.next
                else:
                    self.buckets[index] = node.next
                self.size -= 1
                return
            prev = node
            node = node.next
  
    # Метод изменения емкости таблицы
    def resize(self):
        new_capacity = self.capacity * 2 # увеличиваем емкость вдвое
        new_buckets = [None] * new_capacity # создаем новую таблицу с увеличенной емкостью
        # перехешируем все элементы из старой таблицы в новую
        for i in range(self.capacity):
            node = self.buckets[i]
            while node:
                index = hash(node.key) % new_capacity
                if new_buckets[index]:
                    new_node = new_buckets[index]
                    while new_node.next:
                        new_node = new_node.next
                    new_node.next = Node(node.key, node.value)
                else:
                    new_buckets[index] = Node(node.key, node.value)
                node = node.next
        self.capacity = new_capacity
        self.buckets = new_buckets

ht = HashTable_ceoi4ka()
ht.add("apple", 5)
ht.add("banana", 10)
ht.add("orange", 7)
ht.add("apple", 6)  


start_time = time.time()
ht.get("apple")
print("рандомное цепочкой:--- {0}  ---".format((time.time() - start_time)))
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.

```python
def solve_queens():
    def is_safe(board, row, col):
        # Проверяем вертикаль и горизонталь
        for i in range(8):
            if board[row][i] == 1 or board[i][col] == 1:
                return False
        
        # Проверяем диагонали
        for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        
        for i, j in zip(range(row, -1, -1), range(col, 8)):
            if board[i][j] == 1:
                return False
        
        for i, j in zip(range(row, 8), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        
        for i, j in zip(range(row, 8), range(col, 8)):
            if board[i][j] == 1:
                return False
        
        return True

    def place_queen(board, col):
        if col >= 8:
            return True
        
        for row in range(8):
            if is_safe(board, row, col):
                board[row][col] = 1
                
                if place_queen(board, col + 1):
                    return True
                
                board[row][col] = 0
        
        return False

    # Создаем пустую доску
    board = [[0 for _ in range(8)] for _ in range(8)]
    
    # Пытаемся разместить ферзей
    if place_queen(board, 0):
        # Выводим решение
        for row in board:
            print(' '.join('Q' if cell else '.' for cell in row))
    else:
        print("Решение не найдено")

solve_queens()
```
